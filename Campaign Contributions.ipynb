# Clean Campaign Contribution Data

```python
import pandas as pd
import numpy as np
import json
```

```python
contributions_full = pd.read_csv('FEC filings.csv')
```

```python
contributions_full.head()
```

```python
contributions_full_columns = contributions_full.columns.to_list()
```

```python
contributions_full_columns = contributions_full_columns[1:]
```

```python
contributions_full.drop('election_tp', axis=1, inplace=True)
```

```python
contributions_full.columns = contributions_full_columns
```

```python
contributions_full.head()
```

```python
contributions_full['cand_nm'].value_counts()
```

```python
contributions_bernie = contributions_full[contributions_full.cand_nm == 'Sanders, Bernard']
```

```python
contributions_biden = contributions_full[contributions_full.cand_nm == 'Biden, Joseph R Jr']

```

```python
contributions_df = contributions_bernie.append(contributions_biden)
```

```python
contributions_df.head()
```

```python
contributions_dict = contributions_df.to_dict('records')
```

```python
with open('finances.json', 'w') as file:
    json.dump(contributions_dict, file)
```
